```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time
start_time=time.time()
print(func4([1,2,3,4],[2,4,6,8]))
end_time=time.time()
print("Execution time:", end_time- start_time, "seconds")
#this function count many elements of list a are present in list b using sorting + binary search
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time
start_time=time.time()
print(func8([3,1,2,7,8,6]))
end_time=time.time()
print("Execution time:", end_time- start_time, "seconds")
#a recursive sorting algorithm that sorts an array by repeatedly sorting overlapping two-thirds of it
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(arr):
    total = 0
    for num in arr:
        total += num
    return total
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    max_val = arr[0]
    min_val = arr[0]
    for num in arr[1:]:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num
    return max_val, min_val
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, target):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right)
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
import time
start_time=time.time()
print(binary_search([1, 3, 5, 7, 9,15,16,14,45],9))
end_time=time.time()
print("Execution time:", end_time- start_time, "seconds")
```
